```python
import numpy as np
from matplotlib import pyplot as plt
import math
import heat200Tenv as sys

T_init=sys.get_T_init()
plt.imshow(T_init, cmap='bwr')
```

```python
T_iter=T_init
Tmax=list()
for i in range(int(6*6*300*7)):
    Tmax.append(T_iter.max())
    T_iter=sys.next_T(T_iter)
    if Tmax[-1]>80+sys.T0:
        print(' '+str(i)+' 80')
        break
    pass
Tmax=np.array(Tmax)-sys.T0
print(T_iter.max()-sys.T0)
plt.imshow(T_iter,cmap='bwr')
plt.show()
plt.plot(Tmax)
```

```python
TTT=Tmax
ttt=7*6*6*300*100/60/60/24
TTTconst=TTT[-1]+np.zeros(TTT.size)
ddd=np.linspace(0,ttt,TTT.size)
plt.plot(ddd,TTT,label='$T_{max}$-day')
plt.plot(ddd,TTTconst,label='$T_{final}=$'+str(int(TTT[-1]*100)/100)+'$\degree C$')
plt.legend(fontsize='x-large')
plt.grid(True)
plt.title('$T_{max}$ with Days When \n Number=200 ,$T_{env}=10\degree C$',fontsize='x-large')
plt.xlabel('t / Days')
plt.ylabel('$T_{max}$ / $\degree C$')
plt.savefig('200Tenv10.png',format='png')
```
