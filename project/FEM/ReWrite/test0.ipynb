```python
import FEM as fem
```

```python
node1 = fem.Node(0 , 0 , u = 0 , v = 0)
node2 = fem.Node(1 , 0 , Px = 10000 , v = 0)
sys = fem.SystemTruss([node1 , node2])
sys.Connect(1 , 2 , 200e9 , 0.01)
sys.Generate()
sys.K
```

```python
sys.Solve()
```

```python
sys.SetPrecison(3)
sys.ExportFigure()
```
