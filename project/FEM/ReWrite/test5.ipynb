# 老钱作业

```python
import numpy as np
import FEM as fem
```

```python
P = 10000
a = 1
f = 30e-4
t = 0.1e-2
mu = 0.3
E = 70e9
G = E/2/(1+mu)
```

```python
nodelist = list()
for i in range(6):
    node = fem.Node((5-i)*a , a , Px = 0 , Py = -P)
    nodelist.append(node)
    pass
nodelist[-1].SetBound(u = 0 , v = 0)
for i in range(6):
    node = fem.Node((5-i)*a , 0 , Px = 0 , Py = 0)
    nodelist.append(node)
    pass
nodelist[-1].SetBound(Py = 0 , u = 0)

sys = fem.SystemTruss(nodelist)
```

```python
for i in range(5):
    sys.Connect(i+1 , i+2 , E , f)
    sys.Connect(i+1 , i+7 , E , f)
    sys.Connect(i+7 , i+8 , E , f)
    pass
sys.Connect(6 , 12 , E , f)
for i in range(5):
    sys.SetSheet(12-i , 11-i , 5-i , 6-i , G , t)
    pass
sys.Generate()
```

```python
sys.Solve()
```

```python
print(sys.Report)
```

```python
sys.SetPrecison(3)
```

```python
sys.ExportFigure(sizex = 50 , sizey = 10 , ftsize = 20)
```
